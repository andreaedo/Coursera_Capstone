```
%matplotlib inline 

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
pd.read_csv('https://cocl.us/datascience_survey_data',  index_col = 0)
```

```
df = pd.read_csv('https://cocl.us/datascience_survey_data',  index_col = 0)

df.sort_values(by='Very interested', ascending=False, inplace=True)
df.plot(kind='bar',
        figsize=(20, 8),
        color=['#5cb85c', '#5bc0de', '#d9534f']
       )

plt.title("Percentage of Respondents' Interest in Data Science Areas")

i=0
for index, row in df.iterrows():
 
  plt.annotate(row['Very interested'], xy=(i, row['Very interested']), color="red")
  plt.annotate(row['Somewhat interested'], xy=(i, row['Somewhat interested']), color="blue")
  plt.annotate(row['Not interested'], xy=(i, row['Not interested']), color="black")
  i = i+1

plt.show()
```
